```
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt
```

```
im = imread('fondo.jpg', as_gray=True)
```

```
plt.imshow(im, cmap='gray')
```

```
sd = 1.2
image_stack = []
stack_size = 8
for i in range(stack_size):
  im_noise = im + np.random.normal(0, sd, im.shape)
  # Escalar y convertir a uint8 para guardarlo como PNG
  im_noise = np.clip(im_noise, 0, 1)  # normalizar en rango de 0 a 1
  im_noise = (im_noise * 255).astype(np.uint8) # Escalar de nuevo a uint8

  imsave('im_{}.png'.format(i), im_noise )
  image_stack.append(im_noise)
```

```
image_stack = np.array(image_stack)
image_stack.shape
```

```
plt.figure(figsize = (15,10))
ncols = 5
nrows = 5
for i in range(ncols*nrows):
  plt.subplot(nrows,ncols, i+1)
  plt.axis('off')
  plt.imshow(image_stack[i], cmap='gray')
```
