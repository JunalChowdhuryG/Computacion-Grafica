```python
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt

image_stack = []
stack_size = 1000
for i in range(stack_size):
  im = imread('im_{}.png'.format(i), as_gray = True)
  image_stack.append(im)
```
